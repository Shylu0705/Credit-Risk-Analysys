## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score

%matplotlib inline
```

## Data Loading and Preprocessing

### Load the data

```python
df = pd.read_csv('bankloans.csv')
df.head()
```

### Check for null values

```python
df.isnull().sum()
```

```python
df.value_counts()
```

```python
df = df.dropna()
```

### Exploring Data

Graph of age vs Income

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.lineplot(x='age',y='income',data=df,ax=ax)
```

Graph of age vs ratio of debt to income

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.lineplot(x='age',y='debtinc',data=df,ax=ax)
```

```python
df.drop(['address'], axis = 1)
df.head()
```

## Split train and test sets

```python
df['default'].value_counts()
```

```python
x=df.drop(['default'],axis=1)
y=df['default']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

### Scaling values

```python
sc = StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

## Building Models

### Random Forest Classifier

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
grid_search = GridSearchCV(estimator=rfc, param_grid={
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]},cv=5, scoring='accuracy')
```

```python
grid_search.fit(xtrain, ytrain)
grid_search.best_params_
```

```python
grid_search.score(xtest,ytest)
```

```python
yp = grid_search.predict(xtest)
c= confusion_matrix(ytest,yp)
fig ,ax = plt.subplots(figsize=(20,10))
sns.heatmap(c,ax=ax)
```

### Support vector Classifier

```python
sv = SVC()
sv.fit(xtrain,ytrain)
```

```python
sv.score(xtest,ytest)
```

```python
model = GridSearchCV(sv,{
    'C':[0.05,0.1,0.2,0.4,0.8,1.2,1.8,4.0,7.0],
    'gamma':[0.05, 0.1,0.4,0.8,1.0,2.0,3.0],
    'kernel':['rbf','linear']
},scoring='accuracy',cv=10)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.best_params_
```

```python
model = SVC(C=model.best_params_['C'],gamma=model.best_params_['gamma'],kernel=model.best_params_['kernel'])
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
yp = model.predict(xtest)
c= confusion_matrix(ytest,yp)
fig ,ax = plt.subplots(figsize=(20,10))
sns.heatmap(c,ax=ax)
```

### Logistic Regression

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
yp = lr.predict(xtest)
c= confusion_matrix(ytest,yp)
fig ,ax = plt.subplots(figsize=(20,10))
sns.heatmap(c,ax=ax)
```

## Model evaluation

```python
TN, FP, FN, TP = c.ravel()
print(f"True Negative (TN) : {TN}")
print(f"False Positive (FP): {FP}")
print(f"False Negative (FN): {FN}")
print(f"True Positive (TP) : {TP}")
```

```python
Precision = float(TP)/(TP+FP)*100
Precision
```
